To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
#Making the new 'Year' column with a lambda function
#Taking the respective year of each month of observation
new_df['Year'] = new_df['Month'].apply(lambda t: t.year)

new_df.tail()
```

```python
#Let's get the average prices of each borough for each year

#numeric_only=True just to be sure we only do numeric values
dfg = new_df.groupby(['London_Borough', 'Year']).mean(numeric_only=True)

#sample a few just to be sure it is working
dfg.sample(5)

#Looks good with samples we took
```

```python
#Reset the index on dfg
dfg = dfg.reset_index()

#For confirmation
dfg.head()

#looks good
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#The function
def create_price_ratio(d):
    yr_1998 = float(d['Average_price'][d['Year'] == 1998])
    yr_2018 = float(d['Average_price'][d['Year'] == 2018])
    ratio = [yr_2018/yr_1998]
    return ratio

#We'll test it out on Westminster
create_price_ratio(dfg[dfg['London_Borough'] == 'Westminster'])

#We get roughly 5.35, meaning Westminster is 5.35 time more expensive
##in 2018 than it was 20 years prior
```

```python
#For all boroughs of London

#Initialize an empty dict
final_dict = {}

#The function like the previous function, but for all the boroughs
for bor in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == bor]
    final_dict[bor] = create_price_ratio(borough)

print(final_dict)
```

```python
#Let's consolidate and make this look prettier/easier to read
df_ratios = pd.DataFrame(final_dict)

#Confirmation
df_ratios.head()

#Much better
```

```python
#Let's transpose it to make it even easier to read
df_ratios_T = df_ratios.T

#Reset the index
df_ratios_T = df_ratios_T.reset_index()

#Confirmation
df_ratios_T.head()
```

```python
#Rename the index to 'Borough' and 0 to 2018
##To signify how much higher each borough's price is
###Compared to 20 years prior

df_ratios_T.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)

#confirmation
df_ratios_T.head()

#Good
```

```python
#Let's make a top 10 of the biggest price increases of the boroughs
top10 = df_ratios_T.sort_values(by='2018', ascending=False).head(10)
#to confirm
print(top10)
```

```python
#Let's plot it out for a better visualization
T10_graph = top10[['Borough', '2018']].plot(kind='bar')
T10_graph.set_xticklabels(top10.Borough)
```
